```python
import multiprocessing as mp
import os
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.distance import great_circle as GRC
from mordecai import Geoparser
from dask.distributed import Client, progress
import dask.dataframe as dd
from multiprocessing.pool import ThreadPool as Pool
```

```python
geolocator = Nominatim(user_agent='my_geocoder')
```

```python
def geoparse_location(location):
    try:
        location_data = geo.geoparse(location)
        if location_data:
            return location_data
        else:
            return None
    except GeocoderTimedOut:
        return geoparse_location(location)


def reverse_geoparse_coordinates(lat_long):
    try:
        location_data = geolocator.reverse(lat_long, timeout=10)
        if location_data:
            return location_data
        else:
            return None
    except GeocoderTimedOut:
        return reverse_geoparse_coordinates(lat_long)




def parsingFunction(df):
    # Iterate over each row in the DataFrame without using index
    for index, row in df.iterrows():
        location = row['New_location']
        # Check if the location is a city name or coordinates
        if pd.isnull(location):
            continue
        if "['" in location:
            # Reverse geoparse coordinates
            split_string = location.split(',')
            listSize = len(split_string)

            if listSize == 2:
                lat = split_string[1].strip("']")
                long = split_string[0].strip("['")

                location_data = reverse_geoparse_coordinates((lat, long))

            if listSize == 4:
                # print("list size = 4: ")
                # print(listSize)
                lon1 = split_string[0][2:]
                lat1 = split_string[1]
                lon2 = split_string[2]
                lat2 = split_string[3][:-2]

                coord1 = lat1 + "," + lon1
                coord2 = lat2 + "," + lon2

                distance = GRC(coord1, coord2).km
                # print(distance)
                if distance <= 200:
                    latitude = (float(lat1) + float(lat2)) / 2
                    longitude = (float(lon1) + float(lon2)) / 2

                    coordinates = (latitude, longitude)
                    # print("four: ")
                    # print(coordinates)
                    location_data = reverse_geoparse_coordinates(coordinates)
                else:
                    location = row['user_location']
                    if pd.isnull(location):
                        pass
            # Extract country, state, latitude, and longitude from the geocoded/reverse geocoded location
            if location_data:
                address = location_data.raw.get('address', {})
                df.at[index, 'Country_conf'] = 99
                df.at[index, 'Country'] = address.get('country', '')
                df.at[index, 'State'] = address.get('state', '')
                df.at[index, 'City'] = address.get('city', '')
                df.at[index, 'Latitude'] = location_data.latitude
                df.at[index, 'Longitude'] = location_data.longitude
        else:
            # Geoparse city name
            geonames = geoparse_location(location)
            if geonames is None:
                geonames = {}
            my_dict = {key: value for key, value in enumerate(geonames)}

            if my_dict is None:
                my_dict = {}

            for key in my_dict.keys():
                value = my_dict[key]
                df.at[index, 'Country_conf'] = value['country_conf']
                df.at[index, 'Country'] = value['country_predicted']
                df.at[index, 'City'] = value['word']
                column_name = 'geo'
                if column_name in value:
                    df.at[index, 'State'] = value['geo']['admin1']
                    df.at[index, 'Latitude'] = value['geo']['lat']
                    df.at[index, 'Longitude'] = value['geo']['lon']
    return df
```

```python
if __name__ == '__main__':
    # df = pd.read_csv("E:\Tweet\Dav\Processed_data_Sample\Davnet.csv")
    # Create empty columns for country, state, latitude, and longitude
    df1['Country_conf'] = ''
    df1['Country'] = ''
    df1['State'] = ''
    df1['City'] = ''
    df1['Latitude'] = ''
    df1['Longitude'] = ''

    num_chunks = mp.cpu_count()
    df_split = np.array_split(df1, num_chunks)
    pool = mp.Pool()
    pool.map_async(parsingFunction, df_split)

    pool.close()
    pool.join()

    # df_processed = pd.concat(df_split)

```
